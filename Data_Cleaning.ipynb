```python
# Load up load the dataset from the github repo  to df_loan_data
loan_data_link = 'https://github.com/GabeMaldonado/DS4A_repo/blob/main/DS4A_Project/Datasets/Loan%20Data/loan.xlsx?raw=True'
df_loan_data = pd.read_excel(loan_data_link)
```

```python
# Sample/display 10 random values to get an overall idead of the data
df_loan_data.sample(10)
```

Now that we have the data loaded into a df, we can inspect its datatype as well as counting the `nans`.

```python
# get info about the dataframe: datatype and number of nans
df_loan_data.info(verbose=True, null_counts=True)
```

Looks like we have lots of empty columns. Empty columnns are not relevant for our analyssis as they contain no value information. These empty columns can be safely dropped from the df.

```python
# get the all the names for the empty columns and put them in a list
df_loan_stats = pd.DataFrame(df_loan_data.count(), columns=['counts'], index=None)
empty_cols = list(df_loan_stats[df_loan_stats['counts'] ==0].index)
```

```python
empty_cols
```

```python
# lets drop the empty columns from our data
# create a new df called df_loan_data_no_mt_cols
df_loan_stats_no_mt_cols = df_loan_data.drop(columns=empty_cols)
df_loan_stats_no_mt_cols.info()
```

After dropping the empty columns, we went from 111 columns to 52. We have significanlty reduced the number of columns by dropping columns that were irrelevant. 
We can see that there are missing values in the remaining columns. We can handle the missing values by imputting the mean of the values for each missing data point. This is the preferred approach at this point. However, we might take a different approach during the exploratory data analysis (EDA) stage of the project. 


```python
df_clean = df_loan_stats_no_mt_cols.fillna(df_loan_stats_no_mt_cols.mean()) #took the average and plugged in values for the NaN's
```
